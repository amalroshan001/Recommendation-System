### Calculating IMDB Rating

```python
# Calculate the No.of votes by 80th percentile movie
m= mal_movies['vote_count'].quantile(0.80)
m
```

```python
C=df['vote_average'].mean()
C
```

```python
# function to calculate the IMDB weighted rating for each movie
def weighted_rating(x,m=m,C=C):
    v=x['vote_count']
    R=x['vote_average']
    #compute the weighted rating
    return (v/(v+m)*R)+(m/(m+v)*C)
```

```python
# Considering movies which has runtime between 45 and 300
mal_movies=mal_movies[(mal_movies['runtime']>=45) & (mal_movies['runtime']<=300)]
mal_movies=mal_movies[mal_movies['vote_count']>=m]
```

```python
mal_movies.shape
```

```python
# Applying the function
mal_movies['score']=mal_movies.apply(weighted_rating,axis=1)
```

### Sorting and Output

```python
mal_movies.sort_values(by='score',ascending=False,inplace=True)
```
